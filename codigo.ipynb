```python
import matplotlib.pyplot as plt

# Datos simulados para 6 sprints
sprints = [1, 2, 3, 4, 5, 6]

# Tasa de xito de Envo (en porcentaje)
tasa_exito_envio = [70, 75, 80, 85, 90, 95]

# Nmero de Errores de Validacin
errores_validacion = [20, 15, 10, 8, 5, 2]

# Verificacin de Layout y Diseo (Escala 1-10)
verificacion_layout = [6, 7, 7.5, 8, 8.5, 9]

# Verificacin de Funcionalidad (Escala 1-10)
verificacion_funcionalidad = [5, 6, 7, 8, 9, 9.5]

# Verificacin de Compatibilidad (Escala 1-10)
verificacion_compatibilidad = [4, 5, 6, 7, 8, 9]

# Documentacin de Resultados (Escala 1-10)
documentacion_resultados = [7, 7.5, 8, 8.5, 9, 9.5]

# Satisfaccin del Usuario (Escala 1-10)
satisfaccion_usuario = [5, 6, 7, 8, 9, 10]

# Crear las grficas
fig, axs = plt.subplots(4, 2, figsize=(14, 10))
fig.suptitle('Simulacin de Mtricas de QA a travs de 6 Sprints')

# Tasa de xito de Envo
axs[0, 0].plot(sprints, tasa_exito_envio, marker='o')
axs[0, 0].set_title('Tasa de xito de Envo (%)')
axs[0, 0].set_xlabel('Sprint')
axs[0, 0].set_ylabel('Tasa de xito (%)')

# Nmero de Errores de Validacin
axs[0, 1].plot(sprints, errores_validacion, marker='o', color='red')
axs[0, 1].set_title('Nmero de Errores de Validacin')
axs[0, 1].set_xlabel('Sprint')
axs[0, 1].set_ylabel('Errores')

# Verificacin de Layout y Diseo
axs[1, 0].plot(sprints, verificacion_layout, marker='o', color='green')
axs[1, 0].set_title('Verificacin de Layout y Diseo')
axs[1, 0].set_xlabel('Sprint')
axs[1, 0].set_ylabel('Calidad (1-10)')

# Verificacin de Funcionalidad
axs[1, 1].plot(sprints, verificacion_funcionalidad, marker='o', color='purple')
axs[1, 1].set_title('Verificacin de Funcionalidad')
axs[1, 1].set_xlabel('Sprint')
axs[1, 1].set_ylabel('Calidad (1-10)')

# Verificacin de Compatibilidad
axs[2, 0].plot(sprints, verificacion_compatibilidad, marker='o', color='orange')
axs[2, 0].set_title('Verificacin de Compatibilidad')
axs[2, 0].set_xlabel('Sprint')
axs[2, 0].set_ylabel('Calidad (1-10)')

# Documentacin de Resultados
axs[2, 1].plot(sprints, documentacion_resultados, marker='o', color='brown')
axs[2, 1].set_title('Documentacin de Resultados')
axs[2, 1].set_xlabel('Sprint')
axs[2, 1].set_ylabel('Calidad (1-10)')

# Satisfaccin del Usuario
axs[3, 0].plot(sprints, satisfaccion_usuario, marker='o', color='blue')
axs[3, 0].set_title('Satisfaccin del Usuario')
axs[3, 0].set_xlabel('Sprint')
axs[3, 0].set_ylabel('Satisfaccin (1-10)')

# Quitar el ltimo grfico vaco
fig.delaxes(axs[3, 1])

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
